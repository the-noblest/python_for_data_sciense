### Boston House Prices dataset

```python
target = boston.target
```

```python
y = pd.DataFrame(target, columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred
```

### RandomForestRegressor

### Credit Card Fraud Detection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df['V1'].value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
```

```python
x = df.iloc[:,:-1] # выборка без последнего столбца
y = df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100, stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]
```

```python
gs = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
gs.fit(x_train, y_train)
y_pred = gs.predict(x_test)
```

```python
gs.best_params_
```

```python
y_pred = gs.predict_proba(x_test)
```

```python
y_pred
```

```python
y_pred_proba = y_pred[:,1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
x_corr = x.corr()
x_corr
```

```python
high_corr = x_corr.index[np.abs(x_corr['target']) > 0.5][:-1]
high_corr
```

```python
del x['target']
```

```python
x_2 = pd.DataFrame({
    f'{col}_2': x[col] ** 2 for col in high_corr
})
x_2.head()
```
